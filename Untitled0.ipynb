Diccionario de variables:

Fecha-Date:La fecha de recopilación de datos, tipo: Fecha-object.

ID de persona-Person_ID:Identificador único para cada individuo, tipo: Numérico-int64.

Edad-Age:Edad de la persona (18-60 años), tipo: Numérico-int64.

Género-Gender:Masculino, Femenino u Otro, tipo: Categórico-object.

Hora de inicio del sueño-Sleep Start Time	: Hora en la que la persona se fue a dormir (en formato de 24 horas), tipo: Numérico-float64.

Hora de finalización del sueño-Sleep End Time: Hora en que la persona se despertó (en formato de 24 horas), tipo: Numérico-float64.

Horas totales de sueño-Total Sleep Hours: Duración total del sueño (en horas), tipo: Numérico-float64.

Calidad del sueño-Sleep Quality:Calidad del sueño autoinformada (escala: 1-10), tipo: Categórico-object.

Ejercicio (minutos/día)-Exercise (mins/day)	:Minutos dedicados a hacer ejercicio al día., tipo: Numérico-int64.

Ingesta de cafeína (mg)-Caffeine Intake (mg):Cantidad de cafeína consumida en mg, tipo: Numérico-int64.

Tiempo frente a la pantalla antes de acostarse (minutos)-Screen Time Before Bed (mins):Tiempo dedicado al uso de pantallas antes de dormir, tipo: Numérico-int64.

Horas de trabajo (hrs/día)-Work Hours (hrs/day):Total de horas de trabajo en un día, tipo: Numérico-float64.

Puntuación de productividad-Productivity Score: Puntuación de productividad autoinformada (escala: 1-10), tipo: Numérico-int64.

Puntuación del estado de ánimo-Mood Score: Puntuación del estado de ánimo autoinformado (escala: 1-10), tipo: Numérico-int64.

Nivel de estrés-Stress Level:Nivel de estrés auto-reportado (escala: 1-10), tipo: Numérico-int64.

Vamos a analizar la relación entre variables continuas que pueden influiren en el nivel de prouctivida diario de una persona.

```python
# @title Productividad Promedio vs Horas de Sueño y Calidad del Sueño

# Agrupar los datos por intervalos y calcular la productividad promedio

# Horas de Sueño
intervalos_sueno = np.arange(0, dataset['Total Sleep Hours'].max() + 1, 1)  # Intervalos de 1 hora
medias_productividad_sueno = []

for i in range(len(intervalos_sueno) - 1):
    inicio = intervalos_sueno[i]
    fin = intervalos_sueno[i + 1]
    datos_intervalo = dataset[(dataset['Total Sleep Hours'] >= inicio) & (dataset['Total Sleep Hours'] < fin)]
    if not datos_intervalo.empty:
        media_productividad = datos_intervalo['Productivity Score'].mean()
        medias_productividad_sueno.append(media_productividad)
    else:
        medias_productividad_sueno.append(np.nan)

# Calidad del Sueño
intervalos_calidad_sueno = np.arange(1, 11, 1)  # Intervalos de calidad del sueño (1-10)
medias_productividad_calidad_sueno = []

for i in range(len(intervalos_calidad_sueno)):
    datos_intervalo = dataset[dataset['Sleep Quality'] == intervalos_calidad_sueno[i]]
    if not datos_intervalo.empty:
        media_productividad = datos_intervalo['Productivity Score'].mean()
        medias_productividad_calidad_sueno.append(media_productividad)
    else:
        medias_productividad_calidad_sueno.append(np.nan)

# Crear el gráfico de líneas
plt.figure(figsize=(12, 6))
plt.plot(intervalos_sueno[:-1] + 0.5, medias_productividad_sueno, label='Horas de Sueño', marker='s', color='green')
plt.plot(intervalos_calidad_sueno, medias_productividad_calidad_sueno, label='Calidad del Sueño', marker='^', color='red')
plt.title('Productividad Promedio vs Horas de Sueño y Calidad del Sueño')
plt.xlabel('Valor de la Variable')
plt.ylabel('Productividad Promedio')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# @title Productividad Promedio vs Ingesta de Cafeína

# Agrupar los datos por intervalos y calcular la productividad promedio

intervalos_cafeina = np.arange(0, dataset['Caffeine Intake (mg)'].max() + 50, 50)  # Intervalos de 50 mg
medias_productividad_cafeina = []

for i in range(len(intervalos_cafeina) - 1):
    inicio = intervalos_cafeina[i]
    fin = intervalos_cafeina[i + 1]
    datos_intervalo = dataset[(dataset['Caffeine Intake (mg)'] >= inicio) & (dataset['Caffeine Intake (mg)'] < fin)]
    if not datos_intervalo.empty:
        media_productividad = datos_intervalo['Productivity Score'].mean()
        medias_productividad_cafeina.append(media_productividad)
    else:
        medias_productividad_cafeina.append(np.nan)

# Crear el gráfico de líneas
plt.figure(figsize=(8, 6))
plt.plot(intervalos_cafeina[:-1] + 25, medias_productividad_cafeina, label='Cafeína', marker='*', color='orange')
plt.title('Productividad Promedio vs Ingesta de Cafeína')
plt.xlabel('Ingesta de Cafeína (mg)')
plt.ylabel('Productividad Promedio')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# @title Productividad Promedio vs Tiempo Frente a la Pantalla

# Agrupar los datos por intervalos y calcular la productividad promedio

intervalos_pantalla = np.arange(0, dataset['Screen Time Before Bed (mins)'].max() + 30, 30)  # Intervalos de 30 minutos
medias_productividad_pantalla = []

for i in range(len(intervalos_pantalla) - 1):
    inicio = intervalos_pantalla[i]
    fin = intervalos_pantalla[i + 1]
    datos_intervalo = dataset[(dataset['Screen Time Before Bed (mins)'] >= inicio) & (dataset['Screen Time Before Bed (mins)'] < fin)]
    if not datos_intervalo.empty:
        media_productividad = datos_intervalo['Productivity Score'].mean()
        medias_productividad_pantalla.append(media_productividad)
    else:
        medias_productividad_pantalla.append(np.nan)

# Crear el gráfico de líneas
plt.figure(figsize=(8, 6))
plt.plot(intervalos_pantalla[:-1] + 15, medias_productividad_pantalla, label='Tiempo Frente a Pantalla', marker='s', color='blue')
plt.title('Productividad Promedio vs Tiempo Frente a la Pantalla')
plt.xlabel('Tiempo Frente a Pantalla (minutos)')
plt.ylabel('Productividad Promedio')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# @title Productividad Promedio vs Ejercicio Diario

# Agrupar los datos por intervalos y calcular la productividad promedio

intervalos_ejercicio = np.arange(0, dataset['Exercise (mins/day)'].max() + 30, 30)  # Intervalos de 30 minutos
medias_productividad_ejercicio = []

for i in range(len(intervalos_ejercicio) - 1):
    inicio = intervalos_ejercicio[i]
    fin = intervalos_ejercicio[i + 1]
    datos_intervalo = dataset[(dataset['Exercise (mins/day)'] >= inicio) & (dataset['Exercise (mins/day)'] < fin)]
    if not datos_intervalo.empty:
        media_productividad = datos_intervalo['Productivity Score'].mean()
        medias_productividad_ejercicio.append(media_productividad)
    else:
        medias_productividad_ejercicio.append(np.nan)

# Crear el gráfico de líneas
plt.figure(figsize=(8, 6))
plt.plot(intervalos_ejercicio[:-1] + 15, medias_productividad_ejercicio, label='Ejercicio Diario', marker='o', color='purple')
plt.title('Productividad Promedio vs Ejercicio Diario')
plt.xlabel('Minutos de Ejercicio Diario')
plt.ylabel('Productividad Promedio')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Según los gráficos anteriores, podemos determinar que dormir bien, hacer ejercicio y consumir cafeína de manera moderada ayuda con la productividad diaria.

Dormir bien : La calidad del sueño está fuertemente relacionada con el bienestar y la productividad. Un descanso adecuado permite que las personas se sientan más alertas y concentradas durante el día.

Hacer ejercicio : La actividad física regular no solo mejora la salud física, sino que también se ha demostrado que aumenta la energía y la capacidad de concentración, lo que se traduce en una mayor productividad.

Consumo de cafeína : Aunque la cafeína puede ofrecer un impulso temporal de energía, su consumo excesivo puede interferir con el sueño, lo que a su vez afecta negativamente la productividad. Es importante encontrar un equilibrio en su ingesta para maximizar los beneficios sin comprometer el descanso.
